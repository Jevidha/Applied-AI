```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import HTML
```

```python
def himmelblau(w):
    w1, w2 = w
    return (w1**2 + w2 - 11)**2 + (w1 + w2**2 - 7)**2

def grad_himmelblau(w):
    w1, w2 = w

    df_dw1 = 4*w1*(w1**2 + w2 - 11) + 2*(w1 + w2**2 - 7)
    df_dw2 = 2*(w1**2 + w2 - 11) + 4*w2*(w1 + w2**2 - 7)

    return np.array([df_dw1, df_dw2])
def optimize(optimizer_name, lr=0.01, iterations=200):
    w = np.array([-4.0, 4.0])  # SAME INITIAL POINT
    trajectory = [w.copy()]
    losses = [himmelblau(w)]

    v = np.zeros(2)
    s = np.zeros(2)
    eps = 1e-8
    beta1, beta2 = 0.9, 0.999

    for t in range(1, iterations+1):
        grad = grad_himmelblau(w)

        if optimizer_name == "GD":
            w -= lr * grad

        elif optimizer_name == "SGD":
            noise = np.random.normal(0, 0.1, 2)
            w -= lr * (grad + noise)

        elif optimizer_name == "MiniBatch":
            noise = np.random.normal(0, 0.05, 2)
            w -= lr * (grad + noise)

        elif optimizer_name == "Nesterov":
            mu = 0.9
            v_prev = v.copy()
            v = mu*v - lr*grad_himmelblau(w + mu*v)
            w += v

        elif optimizer_name == "Adagrad":
            s += grad**2
            w -= lr * grad / (np.sqrt(s) + eps)

        elif optimizer_name == "RMSProp":
            s = 0.9*s + 0.1*grad**2
            w -= lr * grad / (np.sqrt(s) + eps)

        elif optimizer_name == "Adam":
            v = beta1*v + (1-beta1)*grad
            s = beta2*s + (1-beta2)*grad**2

            v_hat = v/(1-beta1**t)
            s_hat = s/(1-beta2**t)

            w -= lr * v_hat/(np.sqrt(s_hat)+eps)

        trajectory.append(w.copy())
        losses.append(himmelblau(w))

    return np.array(trajectory), losses
optimizers = ["GD","SGD","MiniBatch","Nesterov","Adagrad","RMSProp","Adam"]

results = {}
for opt in optimizers:
    traj, loss = optimize(opt)
    results[opt] = (traj, loss)
x = np.linspace(-6, 6, 400)
y = np.linspace(-6, 6, 400)
X, Y = np.meshgrid(x, y)
Z = (X**2 + Y - 11)**2 + (X + Y**2 - 7)**2

fig, ax = plt.subplots(figsize=(6,6))
ax.contour(X, Y, Z, levels=50)

lines = {}
for opt in optimizers:
    line, = ax.plot([], [], label=opt)
    lines[opt] = line

ax.legend()

def animate(i):
    for opt in optimizers:
        traj = results[opt][0]
        lines[opt].set_data(traj[:i,0], traj[:i,1])
    return list(lines.values())

anim = animation.FuncAnimation(fig, animate, frames=200, interval=100)
HTML(anim.to_jshtml())
fig, ax = plt.subplots()

lines = {}
for opt in optimizers:
    line, = ax.plot([], [], label=opt)
    lines[opt] = line

ax.set_xlim(0,200)
ax.set_ylim(0,200)
ax.legend()

def animate_loss(i):
    for opt in optimizers:
        loss = results[opt][1]
        lines[opt].set_data(range(i), loss[:i])
    return list(lines.values())

anim2 = animation.FuncAnimation(fig, animate_loss, frames=200, interval=100)
HTML(anim2.to_jshtml())
```
