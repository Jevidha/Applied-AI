```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# 1. Generate random dataset (15 records)
np.random.seed(42)
X = np.random.rand(15, 1) * 10      # Input feature
y = 3 * X.squeeze() + 5 + np.random.randn(15)  # Linear relation with noise

# 2. Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 3. Create MLP Regressor (ANN)
mlp = MLPRegressor(
    hidden_layer_sizes=(10,),
    activation='relu',
    solver='adam',
    max_iter=2000,
    random_state=42
)

# 4. Train the model
mlp.fit(X_train, y_train)

# 5. Make predictions
y_pred = mlp.predict(X_test)

# 6. Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# 7. Plot results
plt.scatter(X, y, color='blue', label='Actual Data')
plt.scatter(X_test, y_pred, color='red', label='Predicted Data')

# Plot regression line
X_line = np.linspace(0, 10, 100).reshape(-1, 1)
y_line = mlp.predict(X_line)
plt.plot(X_line, y_line, color='green', label='ANN Regression Line')

plt.xlabel("Input X")
plt.ylabel("Output y")
plt.title("Linear Regression using ANN (MLP)")
plt.legend()
plt.grid(True)
plt.show()
```
