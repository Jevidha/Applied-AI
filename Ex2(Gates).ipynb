```python
import numpy as np
```

```python
def perceptron(ip_size, lr=0.1):
    return {"lr": lr, "weights": np.array([-0.3, 0.2, 0.2])}
def activation(x):
    if x>= 0:
        return 1
    else:
        return 0
def predict(model, x):
    z = np.dot(x, model["weights"][1:]) - model["weights"][0]
    return activation(z)
def train(model, X, y):
    loop = 0
    while loop<10:
        loop += 1
        ec = 0
        for i in range(len(X)):
            y_pred = predict(model, X[i])
            error = y[i] - y_pred
            model["weights"][1:] += model["lr"] * error * X[i]
            model["weights"][0] -= model["lr"] * error
            if error != 0:
                ec = 1
        if ec == 0:
            break
    print("Epochs:", loop)
    return model
```

```python
X_a = np.array([[0,0],[0,1],[1,0],[1,1]])
y_a = np.array([0,0,0,1])
p_a = perceptron(2)
train(p_a, X_a, y_a)
print("AND Gate:")
for x in X_a:
    print(x, "-->", predict(p_a, x))
```
